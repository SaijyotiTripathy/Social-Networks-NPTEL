### Convergence in Repeated Matrix Multiplication
#### Any vector multiplied repeatedly with the same matrix converges to a constant vector

```python
A= np.mat('1 2;3 4')
v= np.mat('1;1')

for i in range(10):
  z= A*v
  z= z/np.linalg.norm(z)
  print(z)
  v= z
  print("***************")
```

```python
A= np.mat('1 2;3 4')
v= np.mat('3;11')

for i in range(10):
  z= A*v
  z= z/np.linalg.norm(z)
  print(z)
  v= z
  print("***************")
```
