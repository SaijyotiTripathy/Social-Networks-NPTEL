* ### Total possible edges in a graph with n nodes = n*(n-1)/2
* ### Minimum number of edges required to make the graph connected = n log n
* ### Density of graph = ( no of edges ) / ( total possible edges )

* ### Clustering coefficient of a node V in a graph - measure of the degree to which nodes in a graph tend to cluster together. 
### = ( number of links between neighbors of V ) / ( degree of node V )
### = ( number of closed triplets ) / (total number of triplets (open or close) )

```python
def plot_deg_list(G):
  all_degrees= list(dict(G.degree()).values())
  unique_degrees= list(set(all_degrees))

  count_of_degrees=[]

  for i in unique_degrees:
    x= all_degrees.count(i)
    count_of_degrees.append(x)

  #plt.plot(unique_degrees, count_of_degrees, 'bo-')
  plt.loglog(unique_degrees, count_of_degrees, 'bo-')
  plt.xlabel('Degrees')
  plt.ylabel('No of nodes')
  plt.show()

  print('Density : ', nx.density(G))
```

```python
plot_deg_list(G)
print('Clustering Coefficient : ', nx.clustering(G))
print("Average Clustering Coefficient : ", nx.average_clustering(G))
print('Diameter : ', nx.diameter(G))
```

```python
import numpy
import random
```

```python
#Add n number of nodes in the graph
def add_nodes(n):
  G= nx.Graph()
  G.add_nodes_from(range(n))
  return G

#Add one random edge
def add_random_edge(G):
  v1= random.choice(list(G.nodes()))
  v2= random.choice(list(G.nodes()))

  if v1!=v2:
    G.add_edge(v1,v2)
  return G

#Keeps adding random edges in G till it becomes connected
def add_till_connectivity(G):
  while(nx.is_connected(G)==False):
    G= add_random_edge(G)
  return G

#Create an instance of entire process. Input- n, Output- number of edges
def create_instance(n):
  G= add_nodes(n)
  G= add_till_connectivity(G)
  return G.number_of_edges()

#Average it over 100 instances
def create_avg_instance(n):
  list1=[]
  for i in range(100):
    list1.append(create_instance(n))
  return numpy.average(list1)

#Plot tha desired for different number of edges
def plot_connectivity():
  x=[]
  y=[]
  i=10

  while(i<=400):
    x.append(i)
    y.append(create_avg_instance(i))
    i+=10

  plt.xlabel('Number of nodes')
  plt.ylabel('Number of edges required to connect the ')
  plt.plot(x,y)

  x1=[]
  y1=[]
  i=10
  while(i<=200):
    x1.append(i)
    y1.append(i*numpy.log(i))
    i+=10

  plt.plot(x1,y1)
  plt.show()
```

```python
G= add_nodes(10)
print(G.number_of_nodes())
print(nx.is_connected(G))

G= add_till_connectivity(G)
print(G.edges())
print(G.number_of_edges())
print(nx.is_connected(G))
```
