## Simple Branching Process for Modeling Epidemics

### Single Level

Screenshot 2023-04-02 150810.jpg

* #### p = contagiousness of the pathogen at the given node(probability that the child will be affected)
* #### k = number of children the given node has
* #### X = number of nodes infected from the given node

* #### Expected number of secondary infections from the given node : E[X] = k*p

#### Note : All children are independent of eachother

Screenshot 2023-04-02 153714.jpg

#### Expected number of nodes infected from the given node = E[X] = 0.5 * 4 = 2

### Multiple Levels

Screenshot 2023-04-02 154327.jpg

* #### p = contagiousness of the pathogen at the given node(probability that the child will be affected)
* #### k = number of children the given node has
* #### X = number of nodes infected from the given node

#### Note : All children are independent of eachother

* #### Number of people in level 1 = k
#### Expected number of infected people in level 1 : E[X] = k*p

* #### Number of people in level 2 = k^2
#### Expected number of infected people in level 2 : E[X] = (k*p)^2

* #### Number of people in level i = k^i
#### Expected number of infected people in level 2 : E[X] = (k*p)^i

### Basic Reproductive Number
#### Basic Reproductive number = R0 = Expected number of infected people / Number of secondary infections

* #### If R0 < 1 , the infection dies away with probability = 1
#### That means the disease will not become an epidemic

* #### If R0 > 1 , the infection persists  in the network with some positive probability < 1.
#### That means the disease may or maynot become an epidemic

### Knife-edge property
#### Decreasing the R0 value so that it falls below 1 (prevents epidemic)

#### Ways to prevent an epidemic 
* #### Decreasing p - decreasing the contagiousness of the pathogen
* #### Decreasing k - decreasing the neighbors of a person

### Basic Reproductive Number Revisited for Complex Networks

* #### Probability of infection spreading = 2/3

### Using reproductive number concept :

#### R0 = 2 * 2/3 = 4/3
#### => R0 > 1
#### => This proves that the disease persists in the network

### Actual scenario :

#### Probability that the infection will not spread from level 1 to level 2 is = (1 - 2/3)^4 = (1/3)^4 = 1/81
#### Probability that the infection will not spread from level 2 to level 3 is = (1 - 2/3)^4 = (1/3)^4 = 1/81
....
#### Probability that the infection will not spread from level i to level i+1 is = (1 - 2/3)^4 = (1/3)^4 = 1/81

* #### In infinite network the probability that the infection dies away can be calculated as 1. (can be proved by infinite GP formula)

## Percolation Model

#### Percolation models are commonly used to describe the movement of infection through a contact network. Instead of the movement of infections along a contact network, we consider the movement of infectious units along the stages that link hosts to reported infections in humans.

Screenshot 2023-04-04 202719.jpg

* #### In the given diagram, the edges acts as pipes through which infection can flow.
* #### The probability that the infection can flow through that pipe is calculated by tossing a coin where "H" represents "open" pipe and "T" represents "closed" pipe.
* #### The infection can spread from one node to another only if there are open pipes connecting both the nodes.
* #### There's no concept of time here.
